```python
import tensorflow as tf
import pandas as pd 
import matplotlib.pyplot as plt 
```

Getting the data 

```python
#get the data 
data = tf.keras.datasets.boston_housing.load_data(path='boston_housing.npz', test_split=0.2, seed=113)
```

Getting the data from tf as we get the data on form of Tuple of Numpy arrays: (x_train, y_train), (x_test, y_test).So we have to convert it into training an

```python
train_data = data[0]
test_data = data[1]
X_train = pd.DataFrame(train_data[0],columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
Y_train = pd.DataFrame(train_data[1])
X_test = pd.DataFrame(test_data[0],columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
Y_test = pd.DataFrame(test_data[1])
(X_train.shape,Y_train.shape),(X_test.shape,Y_test.shape)
```

```python
X_train.head()
```

```python
X_train.describe()
```

Data Preprocessing 

```python
#Normalize the Data 
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler

ct =make_column_transformer((MinMaxScaler(),['CRIM', 'ZN', 'INDUS', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']))

ct.fit(X_train)

X_train_norm = ct.transform(X_train)
X_test_norm =  ct.transform(X_test)
```

Model

```python
#set seed 
tf.random.set_seed(42)
#Create the model
model = tf.keras.Sequential(name="BostonHouseModel")
model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dense(50))
model.add(tf.keras.layers.Dense(10))
model.add(tf.keras.layers.Dense(1))

#Compile the model
model.compile(loss =tf.keras.losses.mae , optimizer = tf.keras.optimizers.Adam(learning_rate=0.01) , metrics = ["mae"])

#Fit The model
model.fit(X_train,Y_train,epochs = 200,verbose=1)
```

```python
#get Summary of Model 
model.summary()
```

```python
model.evaluate(X_test,Y_test)
```

Normaliozed Data Model

```python
#set seed 
tf.random.set_seed(42)
#Create the model
model = tf.keras.Sequential(name="BostonHouseModelNormData")
model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dense(50))
model.add(tf.keras.layers.Dense(10))
model.add(tf.keras.layers.Dense(1))

#Compile the model
model.compile(loss =tf.keras.losses.mae , optimizer = tf.keras.optimizers.Adam(learning_rate=0.01) , metrics = ["mae"])

#Fit The model
model.fit(X_train_norm,Y_train,epochs = 200,verbose=1)
```

```python
#Evaluate teh Model
model.evaluate(X_test_norm,Y_test)
```
